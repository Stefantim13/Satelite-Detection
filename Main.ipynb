```python
from ultralytics import YOLO
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

Parse the txt files for the YOLO Model

```python
names = [
  'plane',
  'ship',
  'storage-tank',
  'baseball-diamond',
  'tennis-court',
  'basketball-court',
  'ground-track-field',
  'harbor',
  'bridge',
  'large-vehicle',
  'small-vehicle',
  'helicopter',
  'roundabout',
  'soccer-ball-field',
  'swimming-pool'
]
dict = {name: idx for idx, name in enumerate(names)}
os.makedirs("./dataset/labels/train", exist_ok=True)
os.makedirs("./dataset/labels/valid", exist_ok=True)

def convert_format(name):
    cnt = 0
    for file in os.listdir(f"./dataset/labels_start/{name}"):
        if file.endswith(".txt"):
            with open(os.path.join(f"./dataset/labels_start/{name}", file), 'r') as f:
                lines = f.readlines()
            cnt = cnt + 1
            if name == "train":
                print(f"{cnt}/1411")
            else:
                print(f"{cnt}/458")
            # Get image sizes for normalization
            img = file[:-4] + ".png"
            image = cv2.imread(f"./dataset/images/{name}/{img}")
            y, x = image.shape[:2]

            with open(os.path.join(f"./dataset/labels/{name}", file), 'w') as f:
                for line in lines:
                    # For now i will ignore the difficult parameter
                    parts = line.strip().split()
                    if len(parts) == 10:
                        class_id = dict[parts[8]]
                        for i in range(0,8):
                            parts[i] = float(parts[i])
                            if i % 2 == 0:
                                parts[i] /= x
                            else:
                                parts[i] /= y
                            parts[i] = str(parts[i])
                        if class_id < len(names):
                            f.write(f"{class_id} {' '.join(parts[:8])}\n")
                    else:
                        print("Wrong number of parts in line:", line.strip())
        else:
            print(f"File {file} is not a .txt file")

convert_format("train")
convert_format("valid")
```

Train the Yolo Model

```python
model = YOLO("./model/yolo11n-obb.pt")
model.train(
    data="./dataset/data.yaml",
    epochs=5,
    imgsz=640,
    batch=4,
    device="cpu",
    name="find_objects",
)
```

Testing

```python
model = YOLO("./model/detect_objects.pt")

results = model.predict(source="./dataset/images/valid/P0173.png", save=False, imgsz=640)

annotated_frame = results[0].plot()
box = results[0].obb

cv2.imwrite("predicted_output.png", annotated_frame)
cv2.imshow("Prediction", annotated_frame)
cv2.waitKey(0)
cv2.destroyAllWindows()

if box is not None:
    xywhr = box.xywhr.cpu().numpy() 
    conf = box.conf.cpu().numpy()
    cls = box.cls.cpu().numpy()

    for i in range(len(xywhr)):
        cx, cy, w, h, theta = xywhr[i]
        print(f"OBB {i}: Center=({cx:.1f}, {cy:.1f}), Size=({w:.1f}, {h:.1f}), "
              f"Angle={theta:.1f}, Confidence={conf[i]:.2f}, Class={cls[i]}")
else:
    print("No oriented bounding boxes detected.")
```
